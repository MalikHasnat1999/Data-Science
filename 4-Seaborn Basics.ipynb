## Loading Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('marketing_data.csv')
df.head(1)
```

# Line Graph

```python
# Suppose we want to investigate Revenue by Date

sns.lineplot(x="Date", y="Revenue", data=df)

# Arrows band shows the confidence level
# Difficult to investigate
```

```python
# Suppose we want to investigate Revenue by Week_ID

sns.lineplot(x='Week_ID', y='Revenue', data=df)
plt.title('Marketing Graph')

# We Get Data we that we can understand
```

```python
# Adding Category By Promotions

sns.lineplot(x='Week_ID', y='Revenue', data=df, hue='Promo')

# Automaticaly creates Legend and assign color
```

```python
# Adding Category By Promo with "Style"
# Differentiate b/w Lines

sns.lineplot(x="Week_ID", y="Revenue",data=df,hue="Promo",style='Promo')
```

```python
# Changing Size and Confidence Level(ci)

sns.set(rc={'figure.figsize':(8,6)})
sns.lineplot(x='Week_ID',y='Revenue',data=df,hue='Promo',style='Promo',ci=False)
```

```python
## Adding Markers

sns.lineplot(x='Year', y="Revenue",data=df,hue='Day_Name',ci=False, marker='o')
```
