```python
import matplotlib.pyplot as plt
```

```python
views = [120,300,410,560,127,77,893]
days = range(1,8)
plt.plot(days, views, marker='o')
plt.show()
```

## Adding Label and Legend

```python
views = [120,300,41,156,227,777,493]
plt.plot(days, views, marker='o', label="Youtube views")
plt.title("Youtube Views Graph")
plt.xlabel("Days")
plt.ylabel("Views")
plt.legend()

plt.show()
```

## Color, Markers and plot Representation 

```python
views = [12,300,110,560,827,777,893]
plt.plot(days, views,label="twitter views", color='g', marker='o', markerfacecolor='r')
plt.title("My Views")
plt.xlabel("Days")
plt.ylabel("Views")
plt.legend()
plt.show()
```

## Line Style and Width

```python
plt.plot(days, views,label="twitter views",linestyle='-.', marker='o')
plt.title("My Views")
plt.xlabel("Days")
plt.ylabel("Views")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(days, views,label="twitter views",linewidth=3, marker='o')
plt.title("My Views")
plt.xlabel("Days")
plt.ylabel("Views")
plt.legend()
plt.show()
```

## Adding Multiple plot

```python
twitter_views = [171,253,33,221,665,484,773]
facebook_views = [70,53,333,611,562,834,113]
youtube_views = [337,52,623,43,544,184,333]
days = range(1,8)

plt.plot(days, twitter_views, marker='o', label="Twitter")
plt.plot(days, facebook_views, color='b',marker='o', label="facebook")
plt.plot(days, youtube_views, color='r',marker='o', label="Youtube")
plt.title("My Views")
plt.xlabel("Days")
plt.ylabel("Views")
plt.grid(True)
plt.legend()
plt.show()
```
